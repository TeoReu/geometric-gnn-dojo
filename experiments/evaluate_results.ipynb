```python
import pandas as pd
df = pd.read_csv ('results.cvs')
print(df.columns)
```

```python
# Extract the rows with epoch 100 and group by model and rewire
epoch_100_df = df[df["Epoch"] == 100].groupby(['Model', 'Rewire', 'P', 'K'])

# Calculate the mean and standard deviation of the validation MAE over the used seeds
val_mae_mean = epoch_100_df['Val MAE'].mean()
val_mae_std = epoch_100_df['Val MAE'].std()

test_mae_mean = epoch_100_df['Test MAE'].mean()
test_mae_std = epoch_100_df['Test MAE'].std()
# Print the results
print('Validation MAE mean:\n', val_mae_mean)
print('\nValidation MAE std:\n', val_mae_std)

print('Test MAE mean:\n', test_mae_mean)
print('\nTest MAE std:\n', test_mae_std)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data matrix
data = np.array([[60., 50., 55., 85., 100., 85.],
                 [95., 100., 85., 80., 100., 95.],
                 [90., 100., 100., 100., 100., 100.],
                 [95., 100., 95., 100., 100., 100.],
                 [100., 100., 100., 100., 100., 100.]])

# Get the number of layers
num_layers = np.arange(2, 7)

# Plot each line
for i in range(data.shape[1]):
    plt.plot(num_layers, data[:, i], label=f"Line {i+1}")

# Add axis labels and legend
plt.xlabel("Number of Layers")
plt.ylabel("Validation Accuracy")
plt.legend()

# Show the plot
plt.show()
```

```python
def plot_res(d, name, title):
  fig, ax = plt.subplots()

  x = [2, 3,4,5,6,7]
  for key in d.keys():
    l = d[key]

    ax.plot(x, l, label = str(key)+"-hop adj")

  plt.title(title)
  fig.legend(loc='lower right')
  plt.xlabel('Number of layers')
  plt.ylabel('Validation accuracy')


  plt.savefig(name+'.png')
```

```python
data = {0:[60., 50., 55., 85., 100., 85.],
                 1:[95., 100., 85., 80., 100., 95.],
                 2:[90., 100., 100., 100., 100., 100.],
                 3:[95., 100., 95., 100., 100., 100.],
                 4:[100., 100., 100., 100., 100., 100.]}

plot_res(data, "plot1", "N-hop Validation Accuracy")
```
